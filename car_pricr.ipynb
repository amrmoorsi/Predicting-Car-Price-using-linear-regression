```python
encoded_car = pd.get_dummies(df_cars_new,columns=['make','fuel-type','aspiration','body-style','drive-wheels',
                                          'engine-location','engine-type','fuel-system'],drop_first=True)
```

```python
 #delete coulms that highly correalted with each other that cause redundancy in data
```

```python
encoded_car = encoded_car.drop(columns=['make_honda','make_peugot','make_porsche','make_subaru',
                         'engine-location_rear','engine-type_l','engine-type_ohcf','engine-type_rotor',
                    'fuel-system_4bbl','fuel-system_idi'])
```

```python
encoded_car.head()
```

```python
#applying scaling on data to prevent any baising
```

```python
normalized_car = (encoded_car-encoded_car.min())/(encoded_car.max()-encoded_car.min())
```

```python
normalized_car.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
train , test =train_test_split(normalized_car,test_size=0.3)
```

```python
len(train)
```

```python
len(test)
```

```python
train_x = train.drop('price',axis=1)
train_y = train['price']
```

```python
test_x = test.drop('price',axis=1)
test_y = test['price']
```

```python
lr = LinearRegression()
```

```python
lr.fit(train_x,train_y)
```

```python
pred_y=lr.predict(test_x)
```

```python
mean_squared_error(test_y,pred_y)
```
